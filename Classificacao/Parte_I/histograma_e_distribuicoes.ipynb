## DistribuiÃ§Ã£o de probabilidade conjunta

```python
plt.figure(figsize=(15,5))
#plt.hist(df.age, bins=np.arange(0,101,1), rwidth=.8, density=True);
plt.hist2d(df.age, df.MonthlyIncome, bins=30, density=True)


plt.xlabel("Idade", fontsize=16)
plt.ylabel("Probabilidade por Idade", fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```

```python
plt.figure(figsize=(15,5))
#plt.hist(df.age, bins=np.arange(0,101,1), rwidth=.8, density=True);
plt.hist2d(df.age, df.MonthlyIncome, bins=30, density=True)
plt.colorbar()

plt.xlabel("Idade", fontsize=16)
plt.ylabel("Probabilidade por Idade", fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(df.MonthlyIncome, bins=30, density=True)


plt.subplot(1,2,2)
plt.hist2d(df.age, df.MonthlyIncome, bins=30, density=True)
plt.colorbar()

plt.xlabel("Idade", fontsize=16)
plt.ylabel("Probabilidade por Idade", fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(df.MonthlyIncome, bins=30, density=True, orientation="horizontal", rwidth=.8)


plt.subplot(1,2,2)
plt.hist2d(df.age, df.MonthlyIncome, bins=30, density=True)
plt.colorbar()
plt.xlabel("Idade", fontsize=16)
plt.ylabel("Probabilidade por Idade", fontsize=16)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(2,2,1)
plt.hist(df.MonthlyIncome, bins=30, density=True, orientation="horizontal", rwidth=.8)

plt.subplot(2,2,2)
plt.hist2d(df.age, df.MonthlyIncome, bins=30, density=True)
plt.colorbar()
plt.xlabel("Idade", fontsize=16)
plt.ylabel("Probabilidade/Idade", fontsize=16)

plt.subplot(2,2,4)
plt.hist(df.age, bins=30, density=True, rwidth=.8)
```

```python
plt.figure(figsize=(15,2*5))

plt.subplot(2,2,1)
plt.hist(df.MonthlyIncome, bins=30, density=True, orientation="horizontal", rwidth=.8)

plt.subplot(2,2,2)
plt.hist2d(df.age, df.MonthlyIncome, bins=30, density=True)
plt.colorbar()
plt.xlabel("Idade", fontsize=16)
plt.ylabel("Probabilidade/Idade", fontsize=16)

plt.subplot(2,2,4)
plt.hist(df.age, bins=30, density=True, rwidth=.8)
```
