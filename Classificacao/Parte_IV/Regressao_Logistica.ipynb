    - Beta_0 (intercepto) vai deslocar a curva








## O que significa "interpretar" um modelo?

Alguns modelos de negócio precisam conseguir explicar o porque de determinado resultado/ação.
Exemplo: "Por que meu crédito foi negado? O que eu tenho de diferente de fulano?"

Interpretar é saber o quanto **cada variável** impactou para se chegar a um determinado score.

A única coisa que o modelo ajusta são os betas!
