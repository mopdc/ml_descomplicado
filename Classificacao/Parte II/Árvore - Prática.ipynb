### Leitura dos dados

```python
df = pd.read_csv("./dados/dt.csv")
df
```

### Divisão em treino e teste

#### Esses dados não tem uma coluna de tempo, então faremos uma validação Out-of-sample

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, y_treino, y_teste = train_test_split(X,y, test_size=0.30, random_state=61658)
```

```python
display(X_treino)

display(X_teste)
```

```python
y_treino
```

### Definição dos parâmetros de VALIDAÇÃO CRUZADA

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

> Parâmetros GridSearch:

```python
params = {
    'max_leaf_nodes' : [4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048],
    'criterion'      : ['gini', 'entropy']
}
```

> Criando objeto GridSearch:

```python
grid = GridSearchCV(
    DecisionTreeClassifier(random_state=61658), #primeiro argumento: Modelo
    params,                                     #segundo argumento: Parâmetros
    cv=10,                                      #argumento CV: quantidade de folds
    scoring='roc_auc',                          #argumento Scoring: métrica
    verbose=5                                   #argumento Verbose: mostra na tela 
)
```

### Encontrar o melhor modelo nos dados de TREINO por Validação Cruzada:

```python
grid.fit(X_treino, y_treino)
```

### Qual o melhor conjunto de parâmetros que, em média, melhor se ajustou aos nosso dados?

```python
grid.best_params_
```

### Qual foi o melhor score médio?

```python
grid.best_score_
```

### Vamos usar o modelo no conjunto TESTE para ver se está bom

```python
y_pred = grid.predict_proba(X_teste)
display(y_pred)

print("Deixando somente a coluna com os '1'")

y_pred = grid.predict_proba(X_teste)[:,1]
y_pred
```

### Curva ROC

```python
from sklearn.metrics import roc_curve

fpr, tpr, ths = roc_curve(y_teste, y_pred)

# fpr = false positive rate (eixo x)
# tpr = true positive rate (eixo y)
# ths = quais limiares ele testou 

plt.figure(figsize=(8,8))
plt.plot(fpr, tpr)
plt.plot((0,1), (0,1), 'k--')
```

```python
ths
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, ths = roc_curve(y_teste, y_pred)

# fpr = false positive rate (eixo x)
# tpr = true positive rate (eixo y)
# ths = quais limiares ele testou 

plt.figure(figsize=(8,8))
plt.plot(fpr, tpr)
plt.fill_between(fpr, tpr, alpha=.3)
plt.plot((0,1), (0,1), 'k--')
```

### ROC/AUC score

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_teste, y_pred)
```

```python
grid.best_score_
```

### E no conjunto de Treino?

```python
from sklearn.metrics import roc_curve

fpr, tpr, ths = roc_curve(y_teste, y_pred)

plt.figure(figsize=(8,8))
plt.plot(fpr, tpr, label='Teste')
plt.plot((0,1), (0,1), 'k--')

fpr2, tpr2, ths2 = roc_curve(y_treino, grid.predict_proba(X_treino)[:,1])

plt.plot(fpr2, tpr2, label = 'Treino', color='r')

plt.legend(fontsize=16)
```

O conjunto de treino geralmente dá um resultado melhor
