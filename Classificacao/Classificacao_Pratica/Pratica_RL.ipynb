A coluna `Year_Birth` é pouco representativa no modelo de Regressão Logística.

Se fossemos usar uma Árvore, não teríamos problemas, mas a Regressão Logística espera variáveis em uma certa escala, que começa em 0 e vai crescendo sequencialmente.
Ano de nascimento começa e termina em valores muito grandes, com um intervalo muito pequeno entre eles.

Então, uma coisa comum de se fazer nesse tipo de variável, é extrair o que realmente queremos dela: a `Idade`:

```python
df.loc[:, "Idade"] = 2020 - df.Year_Birth  ## A idade que o cliente tinha no momento do contato

df.drop(columns="Year_Birth", axis=1, inplace=True)
```

O mesmo acontece na variável `Dt_Customer`, que é a data de quando o cliente realizou a compra.

Vamos pegar essa variável e vamos criar uma nova coluna com os dias da compra.    

```python
df.loc[:, "Primeira_compra"] = (pd.to_datetime('2020-12-15') - pd.to_datetime(df.Dt_Customer)).dt.days

df.drop(columns="Dt_Customer", axis=1, inplace=True)
```

## Transformar em número algumas colunas que estão em texto:

Coluna `Education` e `Marital_Status`

Por que fazemos depois que separamos nossa base em treino e teste? 
- Porque na base de teste pode ter um tipo que não tem na base de treino.  \
    E é assim que se comporta no mundo real. Quando o modelo não encontrar a categoria no treino, ele informará que aquele campo é contém um dado faltante (dado que ele nunca viu antes)

```python
possiveis_Education = pd.CategoricalDtype(X_treino.Education.unique())

possiveis_Marital_Status = pd.CategoricalDtype(X_treino.Marital_Status.unique())
```

Como vamos impor isso nos nosso dados?

```python
X_treino.loc[:,"Education"] = X_treino.Education.astype(possiveis_Education)
X_teste.loc[:,"Education"]  = X_teste.Education.astype(possiveis_Education)

X_treino.loc[:,"Marital_Status"] = X_treino.Marital_Status.astype(possiveis_Marital_Status)
X_teste.loc[:,"Marital_Status"]  = X_teste.Marital_Status.astype(possiveis_Marital_Status)

X_treino
```

**"Categorical":** só recebe os dados que estão na categoria. Se adicionarmos qualquer valor fora do que está na categoria, ele informará que é um valor faltante. E o modelo que se vire para tratar valores faltantes.

## Tratando os Valores Faltantes:

Isso acontece quando a pessoa não informou.

```python
X_treino_d.isnull().sum()
```

```python
X_teste_d.isnull().sum()
```

## Como vamos suprir esse valor faltante?

Podemos usar um fillna(0) e deixar o valor 0 para quem não informou

Mas podemos usar também outro método

Na regressão logística, quando vamos preencher os valores faltantes, é interessante que esse número faça sentido. Como vamos fitar uma reta, esses valores "nada a ver" podem prejudicar o resultado.

Nesse caso, vamos usar a mediana do salário para preencher essa variável.

```python
valor_para_imputar = X_treino_d.Income.median()

X_treino_d.loc[X_treino_d.Income.isnull(), "Income"] = valor_para_imputar
X_teste_d.loc[X_teste_d.Income.isnull(), "Income"] = valor_para_imputar
```

```python
valor_para_imputar
```

## Agora, temos que fazer a interação das nossas variáveis:

Temos muitas variáveis diferentes, temos que transformar isso de alguma maneira.

### Padronização das variáveis contínuas:

## Opção 1: Escalar as variáveis antes de fazer a interação:

Vai pegar cada variável informada e vai colocar a variável média 0 e desvio padrão 1

A maior parte dos dados estará entre -1 e 1 com média 0

Isso é interessante quando as variáveis não tem tamanho bizarramente altos. Renda é geralmente um exemplo disso, mas nessa base não está assim

```python
X_treino_d.max()
```

```python
from sklearn.preprocessing import StandardScaler 


## Selecionamos as colunas que tem máximo > 10

numericas = [
    'Income',
    'Recency',
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds',
    'NumDealsPurchases',
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases',
    'NumWebVisitsMonth',
    'Idade',
    'Primeira_compra',
]


scalers = {}
X_treino_s = X_treino_d.copy()
X_teste_s = X_teste_d.copy()

for col in numericas:
    scaler = StandardScaler()
    scaler.fit(X_treino_s[[col]]) 
    X_treino_s.loc[:,col] = scaler.transform(X_treino_s[[col]])
    X_teste_s.loc[:,col]  = scaler.transform(X_teste_s[[col]])
    scalers[col] = scaler
    
X_treino_s
```

## Opção 2: Converter em categóricas ordinais:

Vou pegar as colunas numéricas e criar categorias nelas: pegar a renda e criar faixar, por exemplo:

```python
from sklearn.preprocessing import KBinsDiscretizer

scalers_o = {}
X_treino_o = X_treino_d.copy()
X_teste_o = X_teste_d.copy()

for col in numericas:
    scaler = KBinsDiscretizer(n_bins=5, encode="ordinal")
    scaler.fit(X_treino_o[[col]]) 
    X_treino_o.loc[:,col] = scaler.transform(X_treino_o[[col]])
    X_teste_o.loc[:,col]  = scaler.transform(X_teste_o[[col]])
    scalers_o[col] = scaler
    
X_treino_o
```

```python
from sklearn.preprocessing import KBinsDiscretizer

numericas = [
    'Income',
    'Recency',
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds',
    'NumDealsPurchases',
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases',
    'NumWebVisitsMonth',
    'Idade',
    'Primeira_compra',

]

scalers = {}
X_treino_k = X_treino_d.copy()
X_teste_k = X_teste_d.copy()
for col in numericas:
    scaler = KBinsDiscretizer(n_bins=5,encode='onehot-dense',strategy='quantile')
    scaler.fit(X_treino_k[[col]])
    
    novas_colunas = []
    for comeco in scaler.bin_edges_[0][:-1]:
        novas_colunas.append(f'{col}_' + str(comeco))
    
    X_treino_k = X_treino_k.join( pd.DataFrame(scaler.transform(X_treino_k[[col]]),columns=novas_colunas,index=X_treino_k.index) )
    X_teste_k = X_teste_k.join( pd.DataFrame(scaler.transform(X_teste_k[[col]]),columns=novas_colunas,index=X_teste_k.index) )
    
    X_treino_k = X_treino_k.drop(col,axis=1)
    X_teste_k = X_teste_k.drop(col,axis=1)
    scalers[col] = scaler
    
X_treino_k
```

## Correlação

Vamos verificar se temos variáveis correlacionadas

Variáveis muito parecidas "estragam" a Regressão Logística

```python
corr_cat = X_treino_f2.corr("spearman")
corr_cat
```

```python
from matplotlib.colors import TwoSlopeNorm

plt.figure(figsize=(15,15))
plt.imshow(corr_cat,cmap='bwr',norm=TwoSlopeNorm(0.7,vmin=0,vmax=1))
plt.colorbar()

# plt.xticks(range(corr_cat.shape[1]), X_tr_f2.columns,rotation=90, fontsize=16)
# plt.yticks(range(corr_cat.shape[0]), X_tr_f2.columns,rotation=0, fontsize=16);
```

```python
from sklearn.feature_selection import mutual_info_classif

X_treino_f3 = X_treino_f2.copy()
X_teste_f3 = X_teste_f2.copy()
dropped = []

for num1,col1 in enumerate(corr_cat.columns):
    if col1 in dropped:
        continue
    for num2,col2 in enumerate(corr_cat.index):
        if col2 in dropped:
            continue
        if num2 <= num1:
            continue
        if col1==col2:
            continue
        val = corr_cat.loc[col2,col1]
        if val < .7:
            continue
            
        imm1 = mutual_info_classif(X_treino_f3[[col1]],y_treino)/mutual_info_classif(y_treino.values.reshape(-1,1),y_treino)
        imm2 = mutual_info_classif(X_treino_f3[[col2]],y_treino)/mutual_info_classif(y_treino.values.reshape(-1,1),y_treino)
        
        if imm1 > imm2:
            dropped.append(col2.strip())
            print(f'Entre {col1} e {col2}, descartamos {col2}')
        else:
            dropped.append(col1.strip())
            print(f'Entre {col1} e {col2}, descartamos {col1}')

X_treino_f3 = X_treino_f3.drop(dropped,axis=1)
X_teste_f3 = X_teste_f3.drop(dropped,axis=1)
X_treino_f3
```

## Regressão Logística sem Regularização

- A Regressão logística por padrão tem a regularização L2! 

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty='none', max_iter=10_000)
lr.fit(X_treino_f3, y_treino)
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(y_teste, lr.predict_proba(X_teste_f3)[:,1])
auc
```

## Regressão Logística com Regularização L1

## CUIDADO! 

- Para usar a regularização L1, você tem que alterar o argumento `solver` da `LogisticRegression`

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# C = 1/alpha
params = {
    'C' : [0.1,1.10,50]
    
}

lr2 = LogisticRegression(penalty='l1', solver='saga', max_iter=10_000)
grid = GridSearchCV(lr2, params, cv=10, scoring='roc_auc', n_jobs=-1, verbose=10)
```

```python
grid.fit(X_treino_f3, y_treino)
```

## Melhor parâmetro:

```python
grid.best_params_ 
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(y_teste, grid.predict_proba(X_teste_f3)[:,1])
auc
```

### Nosso modelo teve uma AUC muito melhor devido a regularização L1: que é uma regularização que ZERA os betas

```python
print(f'{"Intercepto":40s} : {grid.best_estimator_.intercept_[0]:.1f}')
for col, coef in zip(X_treino_f3.columns, grid.best_estimator_.coef_[0]):
    print(f'{col:40s} : {coef:.3f}')
```

```python
sel_cols = []

print(f'{"Intercepto":40s} : {grid.best_estimator_.intercept_[0]:.1f}')
for col, coef in zip(X_treino_f3.columns, grid.best_estimator_.coef_[0]):
    if coef == 0:
        continue
    sel_cols.append(col)
    print(f'{col:40s} : {coef:.3f}')
```

#### Quando usamos regularização, perdemos a probabilidade das nossas variáveis.

Selecionamos acima as variáveis com os betas != 0 e vamos treinar a Regressão Logística novamente **SEM REGULARIZAÇÃO** com as variáveis selecionadas: 

```python
from sklearn.linear_model import LogisticRegression

lr_sel = LogisticRegression(penalty='none', max_iter=10_000, solver="saga")
lr_sel.fit(X_treino_f3[sel_cols], y_treino)

auc = roc_auc_score(y_teste, lr_sel.predict_proba(X_teste_f3[sel_cols])[:,1])
auc
```

```python
print(f'{"Intercepto":40s} : {lr_sel.intercept_[0]:.1f}')
for col, coef in zip(X_treino_f3.columns, lr_sel.coef_[0]):
    print(f'{col:40s} : {coef:.3f}')
```

```python
lr_sel.predict_proba(X_teste_f3[sel_cols])[:,1]
```

## Mas podemos testar a L2 também! 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

params = {
    'C' : [0.05, 0.1, 1, 10, 50, 100],
}

lr3 = LogisticRegression(penalty='l2', max_iter=10_000, solver="saga")
grid2 = GridSearchCV(lr3, params, cv=10, scoring="roc_auc", n_jobs=-1, verbose=10)
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(y_teste, grid2.predict_proba(X_teste_f3)[:,1])
auc
```

```python
sel_cols = []
print(f'{"Intercepto":40s} : {grid2.best_estimator_.intercept_[0]:.1f}')
for col, coef in zip(X_treino_f3.columns,grid2.best_estimator_.coef_[0]):
    if coef==0:
        continue
    sel_cols.append(col)
    print(f'{col:40s} : {coef:.3f}')
```

#### A Regularização L2 tem bastante dificuldade em zerar os betas. Ela diminui bastante os betas, chega perto de 0, mas não zera.
Podemos usar essa função `np.isclose()` que vai selecionar os betas próximos de 0, com variação de 0.01 para cima ou para baixo:

```python
sel_cols = []
print(f'{"Intercepto":40s} : {grid2.best_estimator_.intercept_[0]:.1f}')
for col, coef in zip(X_treino_f3.columns,grid2.best_estimator_.coef_[0]):
    if np.isclose(coef,0,atol=0.01):
        continue
    sel_cols.append(col)
    print(f'{col:40s} : {coef:.3f}')
```
