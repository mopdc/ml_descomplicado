```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('marketing_campaign.csv', sep=';')
df
```

```python
df.iloc[:,0:20]
```

```python
df.iloc[:,20:]
```

```python
df = df.set_index('ID')
df.loc[:,'Idade'] = 2020 - df.Year_Birth
df.loc[:,'Ultima_compra'] = (pd.to_datetime('2020-12-15') - pd.to_datetime(df.Dt_Customer)).dt.days
df = df.drop(['Year_Birth','Dt_Customer'],axis=1)
df
```

```python
X = df.drop('Response',axis=1)
y = df.Response
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_ts, y_tr, y_ts = train_test_split(X,y,test_size=0.30)
```

```python
df.Education.unique()[1:]
```

```python
possiveis_Education = pd.CategoricalDtype(categories=df.Education.unique())
possiveis_Marital = pd.CategoricalDtype(categories=df.Marital_Status.unique())
possiveis_Education
```

```python
X_tr.loc[:,'Education'] = X_tr.Education.astype(possiveis_Education)
X_ts.loc[:,'Education'] = X_ts.Education.astype(possiveis_Education)

X_tr.loc[:,'Marital_Status'] = X_tr.Marital_Status.astype(possiveis_Marital)
X_ts.loc[:,'Marital_Status'] = X_ts.Marital_Status.astype(possiveis_Marital)

X_tr
```

```python
X_tr_d = pd.get_dummies(X_tr,columns=['Education','Marital_Status'], drop_first=True)
X_ts_d = pd.get_dummies(X_ts,columns=['Education','Marital_Status'], drop_first=True)
X_tr_d
```

```python
X_tr_d.isnull().sum()
```

```python
X_ts_d.isnull().sum()
```

```python
valor_para_inputar = X_tr_d.Income.median()
X_tr_d.loc[X_tr_d.Income.isnull(),'Income'] = valor_para_inputar
X_ts_d.loc[X_ts_d.Income.isnull(),'Income'] = valor_para_inputar
```

```python
valor_para_inputar
```

```python
from sklearn.preprocessing import StandardScaler

numericas = [
    'Income',
    'Recency',
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds',
    'NumDealsPurchases',
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases',
    'NumWebVisitsMonth',
    'Idade',
    'Ultima_compra',
]

scalers = {}
X_tr_s = X_tr_d.copy()
X_ts_s = X_ts_d.copy()
for col in numericas:
    scaler = StandardScaler()
    scaler.fit(X_tr_s[[col]])
    X_tr_s.loc[:,col] = scaler.transform(X_tr_s[[col]])
    X_ts_s.loc[:,col] = scaler.transform(X_ts_s[[col]])
    scalers[col] = scaler
    
X_tr_s
```

# Opção 2: Converter em categóricas ordinais

```python
from sklearn.preprocessing import KBinsDiscretizer

numericas = [
    'Income',
    'Recency',
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds',
    'NumDealsPurchases',
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases',
    'NumWebVisitsMonth',
    'Idade',
    'Ultima_compra',

]

scalers = {}
X_tr_o = X_tr_d.copy()
X_ts_o = X_ts_d.copy()
for col in numericas:
    scaler = KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='quantile')
    scaler.fit(X_tr_o[[col]])
    X_tr_o.loc[:,col] = scaler.transform(X_tr_o[[col]])
    X_ts_o.loc[:,col] = scaler.transform(X_ts_o[[col]])
    scalers[col] = scaler
    
X_tr_o
```

```python
from sklearn.preprocessing import KBinsDiscretizer

numericas = [
    'Income',
    'Recency',
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds',
    'NumDealsPurchases',
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases',
    'NumWebVisitsMonth',
    'Idade',
    'Ultima_compra',

]

scalers = {}
X_tr_k = X_tr_d.copy()
X_ts_k = X_ts_d.copy()
for col in numericas:
    scaler = KBinsDiscretizer(n_bins=5,encode='onehot-dense',strategy='quantile')
    scaler.fit(X_tr_k[[col]])
    
    novas_colunas = []
    for comeco in scaler.bin_edges_[0][:-1]:
        novas_colunas.append(f'{col}_' + str(comeco))
    
    X_tr_k = X_tr_k.join( pd.DataFrame(scaler.transform(X_tr_k[[col]]),columns=novas_colunas,index=X_tr_k.index) )
    X_ts_k = X_ts_k.join( pd.DataFrame(scaler.transform(X_ts_k[[col]]),columns=novas_colunas,index=X_ts_k.index) )
    
    X_tr_k = X_tr_k.drop(col,axis=1)
    X_ts_k = X_ts_k.drop(col,axis=1)
    scalers[col] = scaler
    
X_tr_k
```

```python
from sklearn.preprocessing import PolynomialFeatures

polyfeats = PolynomialFeatures(degree=2, include_bias=False)
polyfeats.fit(X_tr_s)

new_cols = [
    '*'.join(filter(None, [
        f'{col}^{num}' if num!=0 else None for num,col in zip(powers,X_tr_s.columns)
    ])).strip('*')
    for powers in polyfeats.powers_
]

X_tr_f = pd.DataFrame(polyfeats.transform(X_tr_s), columns=new_cols)
X_ts_f = pd.DataFrame(polyfeats.transform(X_ts_s), columns=new_cols)
X_tr_f
```

# Algumas interações não fazem sentido....

```python
const_cols = X_tr_f.columns[ (X_tr_f.max() - X_tr_f.min())==0 ]
const_cols
```

```python
X_tr_f2 = X_tr_f.copy().drop(const_cols,axis=1)
X_ts_f2 = X_ts_f.copy().drop(const_cols,axis=1)
X_tr_f2
```

```python
dup_cols = X_tr_f2.columns[ X_tr_f2.T.duplicated() ]
dup_cols
```

```python
X_tr_f2 = X_tr_f2.drop(dup_cols,axis=1)
X_ts_f2 = X_ts_f2.drop(dup_cols,axis=1)
```

```python
import scipy.stats as ss

# def cramers_v(x, y):
#     confusion_matrix = pd.crosstab(x,y)
#     chi2 = ss.chi2_contingency(confusion_matrix)[0]
#     n = confusion_matrix.sum().sum()
#     phi2 = chi2/n
#     r,k = confusion_matrix.shape
#     phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
#     rcorr = r-((r-1)**2)/(n-1)
#     kcorr = k-((k-1)**2)/(n-1)
#     return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))

corr_cat = X_tr_f2.corr('spearman')
corr_cat
```

```python
from matplotlib.colors import TwoSlopeNorm

plt.figure(figsize=(15,15))
plt.imshow(corr_cat,cmap='bwr',norm=TwoSlopeNorm(0.7,vmin=0,vmax=1))
plt.colorbar()

# plt.xticks(range(corr_cat.shape[1]), X_tr_f2.columns,rotation=90, fontsize=16)
# plt.yticks(range(corr_cat.shape[0]), X_tr_f2.columns,rotation=0, fontsize=16);
```

```python
from sklearn.feature_selection import mutual_info_classif

X_tr_f3 = X_tr_f2.copy()
X_ts_f3 = X_ts_f2.copy()
dropped = []

for num1,col1 in enumerate(corr_cat.columns):
    if col1 in dropped:
        continue
    for num2,col2 in enumerate(corr_cat.index):
        if col1 in dropped:
            break
        if col2 in dropped:
            continue
        if num2 <= num1:
            continue
        if col1==col2:
            continue
        val = corr_cat.loc[col2,col1]
        if val < .7:
            continue
            
        imm1 = mutual_info_classif(X_tr_f3[[col1]],y_tr)/mutual_info_classif(y_tr.values.reshape(-1,1),y_tr)
        imm2 = mutual_info_classif(X_tr_f3[[col2]],y_tr)/mutual_info_classif(y_tr.values.reshape(-1,1),y_tr)
        
        if imm1 > imm2:
            dropped.append(col2.strip())
            print(f'Entre {col1} e {col2}, descartamos {col2}')
        else:
            dropped.append(col1.strip())
            print(f'Entre {col1} e {col2}, descartamos {col1}')

X_tr_f3 = X_tr_f3.drop(dropped,axis=1)
X_ts_f3 = X_ts_f3.drop(dropped,axis=1)
X_tr_f3
```

# CUIDADO!!!!

> - O **PADRÃO** da Logistic Regression é **com regularização** (eu acho um absurdo.....)
> - Se você quer a regressão sem regularização, você deve explicitar

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty='none', max_iter=10_000)
lr.fit(X_tr_f3,y_tr)
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(y_ts, lr.predict_proba(X_ts_f3)[:,1])
auc
```

```python
print(f'{"Intercepto":40s} : {lr.intercept_[0]:.1f}')
for col, coef in zip(X_tr_f3.columns,lr.coef_[0]):
    print(f'{col:40s} : {coef:.1f}')
```

### Vamos tentar usar regularização L1

# CUIDADO!
> - Para usar regularização L1, você tem que mudar o parâmetro `solver` (veja a documentação)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Por escolha dos desenvolvedores, C=1/alpha
params = {
    'C': [0.1,1,10,50]
}

lr2 = LogisticRegression(penalty='l1', max_iter=10_000, solver='saga')
grid = GridSearchCV(lr2, params, cv=10, scoring='roc_auc', n_jobs=-1,verbose=10)
grid.fit(X_tr_f3,y_tr)
```

```python
grid.best_params_
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(y_ts, grid.predict_proba(X_ts_f3)[:,1])
auc
```

```python
print(f'{"Intercepto":40s} : {grid.best_estimator_.intercept_[0]:.1f}')
for col, coef in zip(X_tr_f3.columns,grid.best_estimator_.coef_[0]):
    print(f'{col:40s} : {coef:.3f}')
```

```python
sel_cols = []
print(f'{"Intercepto":40s} : {grid.best_estimator_.intercept_[0]:.1f}')
for col, coef in zip(X_tr_f3.columns,grid.best_estimator_.coef_[0]):
    if coef==0:
        continue
    sel_cols.append(col)
    print(f'{col:40s} : {coef:.3f}')
```

```python
from sklearn.linear_model import LogisticRegression

lr_sel = LogisticRegression(penalty='none', max_iter=10_000,solver='saga')
lr_sel.fit(X_tr_f3[sel_cols],y_tr)

auc = roc_auc_score(y_ts, lr_sel.predict_proba(X_ts_f3[sel_cols])[:,1])
auc
```

```python
print(f'{"Intercepto":40s} : {lr_sel.intercept_[0]:.1f}')
for col, coef in zip(X_tr_f3.columns,lr_sel.coef_[0]):
    print(f'{col:40s} : {coef:.3f}')
```

```python
lr_sel.predict_proba(X_ts_f3[sel_cols])[:,1]
```
